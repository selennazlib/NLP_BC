```python
# import the libraries
import spacy 
import spacy.cli
spacy.cli.download('en_core_web_lg')
nlp = spacy.load('en_core_web_lg')
```

```python
# choose the words then compare, and obtain their vectors
word1 = nlp(u'rainy')
sent1 = nlp(u'Today is too hot.')

print("Word 1 vector\n", word1.vector)
print("\n\nSent 1 vector", sent1.vector)
```

```python
# import spatial and define a cosine_similarity function
from scipy.spatial import distance

def cosine_similarity(vec1, vec2):
  return 1 - distance.cosine(vec1, vec2)
```

```python
# write an expression for vector arithmetic

cat = nlp.vocab['cat'].vector
lion = nlp.vocab['lion'].vector

new_vector = cat + lion
```

```python
# list the top 10 closest vectors
computed_sim = []

for word in nlp.vocab:
  if word.has_vector:
    if word.is_lower:
      if word.is_alpha:
        similarity = cosine_similarity(new_vector, word.vector)
        computed_sim.append((word, similarity))

compoted_sim = sorted(computed_sim, key=lambda item: -item[1])
print([w[0].text for w in computed_sim[: 10]])        
```

```python
# import sentiment intensity analyzer and create an object
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
# write a review
review = "Movie wa horrible. I definitely don't recommend it. It was a huge mistake the take my time on it!!!"
```

```python
# obtain the sid scores for the review
sia.polarity_scores(review)
```

```python
# write a function that takes in a review and returns a score of positive, negative and neutral

def pos_neg_rev(review):
  if sia.polarity_scores(review)['compound'] > 0:
    return 'POS'
  elif sia.polarity_scores(review)['compound'] == 0:
    return 'NEUTRAL'
  else:
    return 'NEG'

print(pos_neg_rev(review))
```
