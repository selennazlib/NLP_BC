```python
import spacy 
import spacy.cli
```

```python
spacy.cli.download("en_core_web_lg")
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
nlp(u'lion').vector
```

```python
nlp(u'.').vector
```

```python
tokens = nlp(u'cloud rain sun thunder')

for token1 in tokens:
  for token2 in tokens:
    print(token1.text, token2.text, token1.similarity(token2)) # cosine similarity
```

```python
tokens = nlp(u'like love pavpav') # I made up the last one
for token1 in tokens:
  print(token1.text, token1.has_vector, token1.vector_norm, token1.is_oov) 
```
